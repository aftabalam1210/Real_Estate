```python
from rapidfuzz import process, fuzz
```

```python
import pandas as pd


# Load your data
  # change this to your CSV path

# Step 1: Lowercase and basic strip
df1['PropertyName'] = df1['PropertyName'].astype(str).str.lower().str.strip()

# Step 2: Get unique names
unique_names = df1['PropertyName'].unique()

# Step 3: Group similar names using fuzzy matching
grouped_names = {}
processed = set()

for name in unique_names:
    if name in processed:
        continue
    # Find all similar names
    matches = process.extract(name, unique_names, scorer=fuzz.token_sort_ratio, score_cutoff=90)
    group_leader = name  # you can also make this more meaningful
    for match_name, score, _ in matches:
        grouped_names[match_name] = group_leader
        processed.add(match_name)

# Step 4: Map all property names to their cleaned group
df1['GroupedName'] = df['PropertyName'].map(grouped_names)

# Step 5: Export
x =df.to_csv("fuzzy_grouped_properties1.csv", index=False)

print("✔️ Grouping complete. Check 'fuzzy_grouped_properties.csv'")
```
